```python
from anyio import Path


df_path= Path('/content/drive/MyDrive/sports.csv')
zip_path = Path('/content/drive/MyDrive/sports_archive.zip')  #the path where the training images are stored
extract_path = Path('/content/drive/MyDrive/sports_archive')  #the path where the training images will be extracted to


df = pd.read_csv(df_path)
```

```python
df.head()
```

Renaming the column names

```python
df.columns = df.columns.str.replace(' ', '_').str.lower()
df.head()
```

```python
datablock = DataBlock(blocks=(ImageBlock, CategoryBlock),
                   splitter=ColSplitter('data_set'),   #based on the column 'data_set' in the dataframe, we are splitting the datas
                   get_x=ColReader('filepaths',pref=extract_path/'/'),   #the source of the data is the path to the images
                   get_y=ColReader('labels'),
                   item_tfms = [Resize(192, method = 'squish')],
                   batch_tfms=aug_transforms()
                   )
```

```python
dls = datablock.dataloaders(df)
dls.show_batch()
```
